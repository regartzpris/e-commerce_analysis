```python
import os
import pandas as pd


# 1. Mendapatkan path ke folder Desktop secara otomatis
folder_path = r'C:\protofolio\ecommerce_analytic\data mentah'

# 3. Membaca file menggunakan pandas
df_customers = pd.read_csv(os.path.join(folder_path, 'customers.csv')).copy()
df_customers.head()
```

```python
#cek duplicate id
print(f"total id duplikat {df_customers['customer_id'].duplicated().sum()}")
```

```python
df_customers.isnull().sum()
```

```python
df_customers.info()
```

```python
#change sign_up date data type to date time

df_customers['signup_date']=pd.to_datetime(df_customers['signup_date'])
df_customers.info()
```

```python
#change datatype country,gender as category to save memory
#print(f"{df_customers['loyalty_tier'].unique().sum()}")
#print(f"{df_customers['acquisition_channel'].unique().sum()}")

col_to_cat = ['country','gender','acquisition_channel','loyalty_tier']
for col in col_to_cat:
    df_customers[col]=df_customers[col].astype('category')

df_customers.info()
```

```python
# save clean data
df_customers.to_csv(r'C:\protofolio\ecommerce_analytic\data bersih\dim_customers_clean.csv', index=False)
print("File sudah tersimpan!")
```

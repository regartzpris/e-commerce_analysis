```python
import os
import pandas as pd


# 1. Mendapatkan path ke folder Desktop secara otomatis
folder_path = r'C:\protofolio\ecommerce_analytic\data mentah'

# 3. Membaca file menggunakan pandas
df_events = pd.read_csv(os.path.join(folder_path, 'events.csv')).copy()
df_events.head()
```

```python
df_events.isnull().sum()
#print(df_events.isnull().sum().mean())
```

```python
#fill product id yang kosong -1 sebagai penanda
df_events['product_id']=df_events['product_id'].fillna(-1)
df_events['product_id'].value_counts()
```

```python
df_events['device_type'].unique()
#fill device null as unknown
df_events['device_type']=df_events['device_type'].fillna('unknown')
df_events['device_type'].value_counts()
```

```python
#change data type ['event_type','device_type','traffic_source','page_category'] as category menghemat memory yg di simpan di memory jadi kode mis 1,1,3
col_to_cat=['event_type','device_type','traffic_source','page_category']
for col in col_to_cat:
    df_events[col] = df_events[col].astype('category')

display(df_events.head())   
df_events.info() 
```

```python
#change timestamp to date for time serie analys an merger table 
df_events['timestamp']=pd.to_datetime(df_events['timestamp'])

df_events['created_at']=df_events['timestamp'].dt.date

display(df_events[['timestamp','customer_id','created_at']].head())

```

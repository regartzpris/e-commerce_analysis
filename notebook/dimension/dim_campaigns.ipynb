```python
import os
import pandas as pd


# 1. Mendapatkan path ke folder Desktop secara otomatis
folder_path = r'C:\protofolio\ecommerce_analytic\data mentah'

# 3. Membaca file menggunakan pandas
df_campaigns = pd.read_csv(os.path.join(folder_path, 'campaigns.csv')).copy()
df_campaigns.head()
```

```python
df_campaigns.isnull().sum()
```

```python
df_campaigns['channel'].unique()
df_campaigns['objective'].unique()
df_campaigns['target_segment'].unique()
```

```python
df_campaigns['start_date']=pd.to_datetime(df_campaigns['start_date'])
df_campaigns['end_date']=pd.to_datetime(df_campaigns['end_date'])

df_campaigns['duration_days'] =df_campaigns['end_date'] - df_campaigns['start_date']

display(df_campaigns[['campaign_id','start_date','end_date','duration_days']].head())
```

```python
display(df_campaigns.head())
```

```python
#  durasi menjadi angka (jumlah hari)
df_campaigns['duration_days_int'] = df_campaigns['duration_days'].dt.days

# Mengelompokkan berdasarkan channel untuk melihat performa rata-rata
channel_performance = df_campaigns.groupby('channel')[['expected_uplift', 'duration_days_int']].mean()


display(channel_performance.round(2))

# Mengurutkan berdasarkan expected_uplift dari tertinggi (descending)
sorted_performance = channel_performance.sort_values(by='expected_uplift', ascending=False)

display(sorted_performance.round(2))
```

```python
import os
import pandas as pd


# 1. Mendapatkan path ke folder Desktop secara otomatis
folder_path = r'C:\protofolio\ecommerce_analytic\data bersih'

# 3. Membaca file menggunakan pandas
df_master = pd.read_csv(os.path.join(folder_path, 'master_clean.csv')).copy()

df_transactions = pd.read_csv(os.path.join(folder_path, 'transactions_clean.csv')).copy()



df_master.head()
```

```python
# Hitung rata-rata/perfotma konversi per channel
df_konversi= df_master.groupby('channel')['event_type'].value_counts(normalize=True)*100
print(df_konversi)
```

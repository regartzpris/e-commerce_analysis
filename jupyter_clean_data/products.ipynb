```python
import os
import pandas as pd


# 1. Mendapatkan path ke folder Desktop secara otomatis
folder_path = r'C:\protofolio\ecommerce_analytic\data mentah'

# 3. Membaca file menggunakan pandas
df_products = pd.read_csv(os.path.join(folder_path, 'products.csv')).copy()
df_products.head()
```

```python
#ck duplicate id df_products['product_id']

duplicated_id=df_products[df_products.duplicated('product_id',keep=False)]
print(f"total duplikat {len(duplicated_id)}")
```

```python
df_products.info()
```

```python
#change laundh_date to date tie
df_products['launch_date']=pd.to_datetime(df_products['launch_date'])
df_products.info()
```

```python
df_products['category'].value_counts()
#df_products['brand'].value_counts()

#change category and brand data type as category
col_to_cat= df_products[['category','brand']].astype('category')
for col in col_to_cat:
    df_products[col]= df_products[col].astype('category')

df_products.info()
```
